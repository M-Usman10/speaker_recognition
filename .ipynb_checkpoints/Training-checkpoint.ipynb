```python
import numpy as np
import tensorflow as tf
```

### Loading the training dataset which was generated by preprocessing.ipynb

```python
X=np.load('final_x.npy')
Y=np.load('final_y.npy')
print (X.shape,Y.shape)
```

### Splitting the dataset into train test

```python
# set seed for numpy and tensorflow
# set for reproducible results
seed = 5
np.random.seed(seed)
tf.set_random_seed(seed)
```

```python
def train_test_split(X,Y,test_size):
    D=np.c_[X,Y]
    np.random.shuffle(D)
    Dtrain=D[:int(X.shape[0]*(1-test_size))-int(X.shape[0]*test_size)]
    Dval=D[int(X.shape[0]*(1-test_size))-int(X.shape[0]*test_size):int(X.shape[0]*(1-test_size))]
    Dtest=D[int(X.shape[0]*(1-test_size)):]
    Xtrain=Dtrain[:,:-1]
    Ytrain=Dtrain[:,-1]
    Xval=Dval[:,:-1]
    Yval=Dval[:,-1]
    Xtest=Dtest[:,:-1]
    Ytest=Dtest[:,-1]
    return Xtrain,Xtest,Xval,Ytrain,Ytest,Yval
#Val and test size will be same
Xtrain,Xtest,Xval,Ytrain,Ytest,Yval=train_test_split(X,Y,test_size=0.15)
print (Xtrain.shape,Xtest.shape,Xval.shape,Ytrain.shape,Ytest.shape,Yval.shape)
```

### Helping class for minibatch optimizer

```python
class Data:
    def __init__(self,x,y,size):
        self.i=-1
        self.Data=x
        self.X=list(range(x.shape[0]))
        self.Y=y
        self.size=size
        self.ite=int(np.ceil(float(x.shape[0])/size))
    def next_batch(self):
        self.i+=1
        self.i=self.i%self.ite
        if (self.i==0):
            np.random.shuffle(self.X)
        return self.Data[self.X[self.i*(self.size):(self.i+1)*self.size]],self.Y[self.X[self.i*(self.size):(self.i+1)*self.size]]
```

```python
def acc(a,b):
    return np.mean(a==b)
```

## Building the tensorflow training model

```python
weights=tf.Variable(np.random.randn(X.shape[1],1),dtype=tf.float32)
x=tf.placeholder(dtype=tf.float32,shape=(None,X.shape[1]))
target=tf.placeholder(dtype=tf.float32,shape=(None,1))
b=tf.Variable(0,dtype=tf.float32)
q_polynomial=tf.matmul(x,weights)+b
lambda_=1e-16
loss=tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=q_polynomial, labels=target))+lambda_*tf.nn.l2_loss(b)+lambda_*tf.nn.l2_loss(weights)
learning_rate = 0.1
batch_size = 1000
iter_num = 1000
D=Data(Xtrain,Ytrain,batch_size)
optimizer=tf.train.AdamOptimizer(learning_rate)
train_op=optimizer.minimize(loss)
prediction = tf.round(tf.sigmoid(q_polynomial))
init = tf.global_variables_initializer()
```

### Training

```python
with tf.Session() as sess:
    sess.run(init)
    for i in range(iter_num):
        mini_x,mini_y=D.next_batch()
        sess.run(train_op,feed_dict={x:mini_x,target:mini_y.reshape(-1,1)})
        if (i%50==0):
            print ('Iteration:',i,' Accuracy:',acc(sess.run(prediction,feed_dict={x:Xtrain,target:Ytrain.reshape(-1,1)}).flatten(),Ytrain))
    w=sess.run(weights)
```

### Saving model to hard drive

```python
np.save('Weights.npy',w)
```
